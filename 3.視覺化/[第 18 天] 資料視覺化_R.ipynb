# 基本的圖形，包括：

+ 直方圖（Histogram）
+ 散佈圖（Scatter plot）
+ 線圖（Line plot）
+ 長條圖（Bar plot）
+ 盒鬚圖（Box plot）

## 直方圖（Histogram）

R 語言

使用 hist() 函數。

```r
normal_samples <- runif(100000) # 生成 100000 組標準常態分配（平均值為 0，標準差為 1 的常態分配）隨機變數
uniform_samples <- rnorm(100000) # 生成 100000 組介於 0 與 1 之間均勻分配隨機變數

hist(normal_samples)
hist(uniform_samples)
```

## 散佈圖（Scatter plot）

R 語言

使用 plot() 函數。

```r
plot(cars$speed, cars$dist)
```

## 線圖（Line plot）

R 語言

使用 plot() 函數，指定參數 type = "l"。

```r
plot(cars$speed, cars$dist, type = "l")
```

## 長條圖（Bar plot）

R 語言

使用 barplot() 函數。

```r
barplot(table(mtcars$cyl))
```

## 盒鬚圖（Box plot）

R 語言

使用 boxplot() 函數。

```r
normal_samples <- runif(100000) # 生成 100000 組標準常態分配（平均值為 0，標準差為 1 的常態分配）隨機變數
boxplot(normal_samples)
```
